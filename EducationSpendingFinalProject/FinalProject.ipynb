```python
#import libraries as needed
import pandas as pd
```

```python
#import database
df = pd.read_csv('states_all.csv')
print(df.head())
print(df.tail())
print(df.info())
```

```python
#Check for duplicate data and remove if necessary
df = df.drop_duplicates()
print(df.info())
```

```python
#Pull out data to be used and remove any rows with missing data
working_df = df[['STATE', 'YEAR', 'TOTAL_EXPENDITURE', 'GRADES_ALL_G', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE']]
print(working_df.info())
clean_df = working_df.dropna(how = 'any')
print(clean_df.info())
```

```python
#EDA
spending_per_pupil = (clean_df['TOTAL_EXPENDITURE'] / clean_df['GRADES_ALL_G'])
#print(spending_per_pupil.head())

#descriptive stats
#mean spending_per_pupil for each state
mean_spending_per_pupil = spending_per_pupil.mean()
#median spending_per_pupil for each state
median_spending_per_pupil = spending_per_pupil.median()
#range spending_per_pupil for each state
min_spending_per_pupil = spending_per_pupil.min()
max_spending_per_pupil = spending_per_pupil.max()

print(mean_spending_per_pupil)
print(median_spending_per_pupil)
print(min_spending_per_pupil)
print(max_spending_per_pupil)

```

```python
#mean, median and range of 4th grade scores
mean_4_reading = clean_df.AVG_READING_4_SCORE.mean()
mean_4_math = clean_df.AVG_MATH_4_SCORE.mean()

median_4_reading = clean_df.AVG_READING_4_SCORE.median()
median_4_math = clean_df.AVG_MATH_4_SCORE.median()

min_4_reading = clean_df.AVG_READING_4_SCORE.min()
min_4_math = clean_df.AVG_MATH_4_SCORE.min()

max_4_reading = clean_df.AVG_READING_4_SCORE.max()
max_4_math = clean_df.AVG_MATH_4_SCORE.max()

#mean, median and range of 8th grade scores
mean_8_reading = clean_df.AVG_READING_8_SCORE.mean()
mean_8_math = clean_df.AVG_MATH_8_SCORE.mean()

median_8_reading = clean_df.AVG_READING_8_SCORE.median()
median_8_math = clean_df.AVG_MATH_8_SCORE.median()

min_8_reading = clean_df.AVG_READING_8_SCORE.min()
min_8_math = clean_df.AVG_MATH_8_SCORE.min()

max_8_reading = clean_df.AVG_READING_8_SCORE.max()
max_8_math = clean_df.AVG_MATH_8_SCORE.max()

print(mean_4_reading)
print(mean_4_math)
print(median_4_reading)
print(median_4_math)
print(min_4_reading)
print(min_4_math)
print(max_4_reading)
print(max_4_math)

print(mean_8_reading)
print(mean_8_math)
print(median_8_reading)
print(median_8_math)
print(min_8_reading)
print(min_8_math)
print(max_8_reading)
print(max_8_math)
```

```python
#correlations between spending_per_pupil and test scores
from scipy.stats import pearsonr

spending_scores_4_rdg_corr, p = pearsonr(spending_per_pupil, clean_df.AVG_READING_4_SCORE)
print(spending_scores_4_rdg_corr)

spending_scores_4_math_corr, p = pearsonr(spending_per_pupil, clean_df.AVG_MATH_4_SCORE)
print(spending_scores_4_math_corr)

spending_scores_8_rdg_corr, p = pearsonr(spending_per_pupil, clean_df.AVG_READING_8_SCORE)
print(spending_scores_8_rdg_corr)

spending_scores_8_math_corr, p = pearsonr(spending_per_pupil, clean_df.AVG_MATH_8_SCORE)
print(spending_scores_8_math_corr)
```

```python
#Line graph of spending_per_pupil and test scores over time

from matplotlib import pyplot as plt

plt.scatter(clean_df.AVG_READING_4_SCORE, spending_per_pupil)
plt.xlabel('4th grade reading scores')
plt.ylabel('Spending per Pupil')
plt.title('Spending per Pupil and 4th Grade Reading Scores')
plt.show()
plt.clf()

plt.scatter(clean_df.AVG_MATH_4_SCORE, spending_per_pupil)
plt.xlabel('4th grade math scores')
plt.ylabel('Spending per Pupil')
plt.title('Spending per Pupil and 4th Grade Math Scores')
plt.show()
plt.clf()

plt.scatter(clean_df.AVG_READING_8_SCORE, spending_per_pupil)
plt.xlabel('8th grade reading scores')
plt.ylabel('Spending per Pupil')
plt.title('Spending per Pupil and 8th Grade Reading Scores')
plt.show()
plt.clf()

plt.scatter(clean_df.AVG_MATH_8_SCORE, spending_per_pupil)
plt.xlabel('8th grade math scores')
plt.ylabel('Spending per Pupil')
plt.title('Spending per Pupil and 8th Grade Math Scores')
plt.show()
plt.clf()

```

```python
#MD changes in spending and test scores over time
md_spending = clean_df.TOTAL_EXPENDITURE[(clean_df.STATE == 'MARYLAND')]
md_enrollment = clean_df.GRADES_ALL_G[(clean_df.STATE == 'MARYLAND')]
md_years = clean_df.YEAR[(clean_df.STATE == 'MARYLAND')]
md_4_reading = clean_df.AVG_READING_4_SCORE[(clean_df.STATE == "MARYLAND")]
md_8_reading = clean_df.AVG_READING_8_SCORE[(clean_df.STATE == "MARYLAND")]
md_4_math = clean_df.AVG_MATH_4_SCORE[(clean_df.STATE == "MARYLAND")]
md_8_math = clean_df.AVG_MATH_8_SCORE[(clean_df.STATE == "MARYLAND")]
md_spending_per_pupil = md_spending/md_enrollment

print(md_spending_per_pupil)
print(md_4_reading)
print(md_4_math)
print(md_8_reading)
print(md_8_math)
```

```python
#MD correlations between spending_per_pupil and test scores
md_spending_scores_4_rdg_corr, p = pearsonr(md_spending_per_pupil, md_4_reading)
print(md_spending_scores_4_rdg_corr)

md_spending_scores_4_math_corr, p = pearsonr(md_spending_per_pupil, md_4_math)
print(md_spending_scores_4_math_corr)

md_spending_scores_8_rdg_corr, p = pearsonr(md_spending_per_pupil, md_8_reading)
print(md_spending_scores_8_rdg_corr)

md_spending_scores_8_math_corr, p = pearsonr(md_spending_per_pupil, md_8_math)
print(md_spending_scores_8_math_corr)
```

```python
# Left plot
plt.figure(figsize=(15, 8))
ax1 = plt.subplot(1, 3, 1)
plt.plot(md_years, md_spending_per_pupil)
ax1.set_title('MD Spending Per Pupil')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Expenditure')
plt.xticks(rotation=30)

# Center plot
ax2 = plt.subplot(1, 3, 2)
plt.plot(md_years, md_4_reading)
ax2.set_title('4th Grade Reading Scores')
ax2.set_xlabel('Year')
ax2.set_ylabel('Scores')

#Right plot
ax3 = plt.subplot(1, 3, 3)
plt.plot(md_years, md_4_math)
ax3.set_title('4th Grade Math Scores')
ax3.set_xlabel('Year')
ax3.set_ylabel('Scores')

plt.subplots_adjust(wspace=0.5)
plt.xticks(rotation=30)
```

In MD, scores increased as spending increased but a decrease in funding also resulted in a decrease in future scores - kids may not have gotten skills needed during decreased funding period that lead to future lower scores/negative outcomes

Nationally, Pearson r scores > 0.3 indicates that there is a positive linear relationship between amount spent per pupil and test scores, but it is inconsistent.  Overall, there seems to be a 'sweet spot' where increased spending results in stronger test scores, but then tapers off so that some of the states spending the most money per pupil are not reporting the highest test scores.  
One question this raises is whether or not the higher spending is due to increased spending for students with disabilities, which may be a positive for those states but not be reflected in assessment scores.
